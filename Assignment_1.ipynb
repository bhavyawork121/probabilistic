```python
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(42)
data = np.random.beta(a=5, b=2, size=1000)  # Skewed to the left
data = data * 100  # Scale for visualization
mean_val = np.mean(data)
median_val = np.median(data)
counts, bins = np.histogram(data, bins=30)
mode_val = bins[np.argmax(counts)]
plt.hist(data, bins=30, edgecolor="black", alpha=0.7)
plt.axvline(mean_val, color="red", linestyle="--", linewidth=2, label=f"Mean ≈ {mean_val:.1f}")
plt.axvline(median_val, color="green", linestyle="-.", linewidth=2, label=f"Median ≈ {median_val:.1f}")
plt.axvline(mode_val, color="blue", linestyle=":", linewidth=2, label=f"Mode ≈ {mode_val:.1f}")

plt.title("Left-Skewed Distribution (Histogram)")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.legend()
plt.show()
```

```python
#A population parameter is a numerical value that describes a characteristic of an entire population.
#It is usually fixed but unknown, since measuring the whole population is often impractical.

#A sample statistic is a numerical value that describes a characteristic of a sample drawn from the population.
#It is used to estimate the corresponding population parameter and can vary from sample to sample.

import numpy as np
np.random.seed(42)
population = np.random.normal(loc=70, scale=10, size=5000) 
# Population parameter (true mean)
population_mean = np.mean(population)
sample = np.random.choice(population, 100, replace=False)
# Sample statistic (sample mean)
sample_mean = np.mean(sample)
population_mean, sample_mean

#Population parameter (true mean of all 5000 students) = 70.06
#Sample statistic (mean of 100 randomly chosen students) = 70.94
#The population mean (µ) is the fixed value that describes everyone.
# The sample mean (x̄) is an estimate of µ, based on just part of the population.
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import collections

data = ['F', 'H', 'O', 'F', 'F', 'O', 'H', 'F', 'O', 'F', 'H', 'F', 'F', 'F', 'O', 'H', 'O', 'F', 'H', 'F']

frequency_counts = collections.Counter(data)

df = pd.DataFrame(list(frequency_counts.items()), columns=['Learning Mode', 'Frequency'])

df['Learning Mode'] = df['Learning Mode'].map({'F': 'In-Person (F)', 'H': 'Hybrid (H)', 'O': 'Online (O)'})

df = df.sort_values(by='Frequency', ascending=False).reset_index(drop=True)

print("(a) Frequency Table:")
print(df)

plt.figure(figsize=(8, 6))

bars = plt.bar(df['Learning Mode'], df['Frequency'], color=['#4CAF50', '#FFC107', '#2196F3'])

plt.xlabel("Preferred Learning Mode")
plt.ylabel("Number of Students")
plt.title("Preferred Learning Modes of 20 Students")
plt.ylim(0, 12)

for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2.0, yval, int(yval), va='bottom', ha='center')

plt.savefig("learning_modes_bar_chart.png")
print("\n(b) Bar chart saved as 'learning_modes_bar_chart.png'")

most_popular_freq = df.loc[0, 'Frequency']
total_students = sum(df['Frequency'])
percentage = (most_popular_freq / total_students) * 100

print(f"\n(c) The most popular mode is preferred by {percentage:.0f}% of students.")

#A histogram is not a suitable graph for this type of data because the data is categorical, not numerical.
# Bar charts are designed to compare the frequencies of distinct, separate categories, like "Online," "Hybrid," and "In-Person."
# Histograms, on the other hand, are used to display the distribution of continuous numerical data, such as exam scores or student heights, by grouping them into intervals (or "bins").
```

```python
p_disease = 0.01
p_positive_given_disease = 0.95
p_positive_given_no_disease = 0.02
p_no_disease = 1 - p_disease
p_true_positive = p_positive_given_disease * p_disease
p_false_positive = p_positive_given_no_disease * p_no_disease
p_positive = p_true_positive + p_false_positive
p_disease_given_positive = p_true_positive / p_positive
print(f"The probability of a person having the disease given a positive test is: {p_disease_given_positive:.4f}")
print(f"As a percentage, this is approximately: {p_disease_given_positive:.2%}")
```
